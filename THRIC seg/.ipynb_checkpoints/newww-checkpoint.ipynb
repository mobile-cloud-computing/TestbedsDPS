```python
import numpy as np
import cv2
import glob
import ntpath
import os
from PIL import Image
```

```python
def flatten(seq):
  for el in seq:
    if isinstance(el, list):
      yield from flatten(el)
    else:
      yield el
```

```python
filelist=glob.glob('./thermal2/*.txt')
```

```python
for i in filelist:
    print(i)
    values=[]
    with open(i) as f:
        for line in f:
            linelist=line.split(" ")
            lineint=[int(i) for i in linelist]
            values.append(lineint)
        
    print(np.mean(values))
    white=max(flatten(values))
    black=min(flatten(values))
    if black!=white:
        for index,line in enumerate(values):
            for ind,v in enumerate(line):
                values[index][ind]=int((v-black)/(white-black)*255)
        im = np.array(values, dtype = np.uint8)
        print(im)

        vmin=30300
        vmax=31300
        vmint=int((vmin-black)/(white-black)*255)
        vmaxt=int((vmax-black)/(white-black)*255)

        thresh=29800
        thresht=int((thresh-black)/(white-black)*255)
        #res2 = cv2.inRange(im, 200, 255)
        res3 = cv2.inRange(im, 150, 255)
        #res4 = cv2.inRange(im, 100, 255)
        

        #im_v = cv2.hconcat([im,res2,res3,res4])
        #cv2.imshow('image',im_v)
        cv2.imshow('image',res3)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
        filename=ntpath.basename(i)
        pre=filename.split(".")[0]
        fi='./threshold-thermal2/'+pre+".png"
        cv2.imwrite(fi,res3)
        img = cv2.imread('./threshold-thermal2/'+pre+".png",0)

            
```
